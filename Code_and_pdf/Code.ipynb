```python
#imported random to help with the deletion of at least 2 random characters from a string
import random

#Question number 5a: takes a string, removes 2 characters, reverses the remaining string, then prints it
def reversed():
  #asks user for a string
  input_string = list(input("Enter the string: "))

  #deletes two random characters from the string
  del input_string[random.randint(0, len(input_string) - 1)]
  del input_string[random.randint(0, len(input_string) - 1)]

  #reverses the order of the remaining string
  reversed = ''.join(input_string[::-1])

  #returns the reversed string
  return reversed

#Question number 5b: takes 2 numbers and performs arithmetic on them
def arithmetic():
  #asks the user to input two numbers
  num1 = float(input("Enter the first number: "))
  num2 = float(input("Enter the second number(non-zero): "))

  #does all the arithmetic: addition, subtraction, multiplication, and division
  addition = num1 + num2
  subtraction = num1 - num2
  multiplication = num1 * num2
  #Added a cannot divide by 0
  division = num1 / num2 if num2 != 0 else print("Cannot divide by 0")

  #prints the results of all of the arithmetic
  print(f"Add: {addition}")
  print(f"Subtract: {subtraction}")
  print(f"Multiply: {multiplication}")
  print(f"Divide: {division}")
  print()

#Question number 6: Takes a sentence and replaces all occurences of python with pythons
def python_to_pythons():
  #asks the user to enter a string
  sentence = str(input("Please enter a sentence with the word python in it: ")).split(' ')
  #creates a count in order to change python to pythons later
  count = 0
  #for loop that looks at each word in the sentence
  for word in sentence:
    if word == 'python':
      #if python is found, change it to pythons
      sentence[count] = 'pythons'
    count += 1

  #Prints the resulting sentence as a sentence and not a list
  print(' '.join(word for word in sentence))
  print()

#Question number 7: uses the grading criteria of this class to assign grades based on a float percentage grade
def grading():
  #asks user for their grade
  grade = float(input("Please enter the grade you have: "))
  #bunch of if statements that will place your grade into a letter grade
  if grade >= 90.0:
    print("You have an A in the class")
  elif grade < 90.0 and grade > 80.0:
    print("You have a B in the class")
  elif grade < 90.0 and grade > 80.0:
    print("You have a B in the class")
  elif grade < 80.0 and grade > 70.0:
    print("You have a C in the class")
  elif grade < 70.0 and grade > 60.0:
    print("You have a D in the class")
  elif grade < 60.0:
    print("You have an F in the class")
  print()

#Question number 8: takes a list and prints out all of the types of each element
def type_of_elements():
  #list given in question
  x = [23, 'Python', 23.98]
  #gets the type of each element
  types = [type(element) for element in x]
  #prints the original list and then prints the types
  print(x)
  print(types)
  print()

#Question 9
def sets():
  #The given sets in the question
  IT_companies = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}
  A = {19, 22, 24, 20, 25, 26}
  B = {19, 22, 20, 25, 26, 24, 28, 27}
  ages = [22, 19, 24, 25, 26, 24, 25, 24]

  #finds the length of IT_companies and then adds twitter
  print(f"Length of IT companies: {len(IT_companies)}")
  print(IT_companies)
  IT_companies.add('Twitter')
  print(f"Length of IT companies: {len(IT_companies)}")
  print(IT_companies)

  #adds multiple companies to IT_companies
  IT_companies.update(['Instagram', 'Snapchat', 'Spotify'])
  print(f"Length of IT companies: {len(IT_companies)}")
  print(IT_companies)

  #removes a company from IT_companies
  IT_companies.remove('Snapchat')
  print(f"Length of IT companies: {len(IT_companies)}")
  print(IT_companies)
  print()

  #The difference between discard and remove is that remove will raise an error if there is nothing with that value in the
  #set, whereas discard does not raise an error

  #joining A and B
  AUB = A.union(B)
  print("Joining of A and B:")
  print(AUB)
  print()

  #finding the intersection of A and B
  A_Intersect_B = A.intersection(B)
  print("Intersection of A and B: ")
  print(A_Intersect_B)
  print()

  #Is A a subset of B
  print(f"Is A a subset of B: {A.issubset(B)}")
  print()

  #Are A and B disjoint sets?
  print("Are A and B disjoint sets?")
  print(A.isdisjoint(B))
  print()

  #Symmetric difference between A and B
  print("Symettric difference between A and B:")
  print(A.symmetric_difference(B))
  print()

  #Delete the sets A, B, and IT_companies completely
  del A
  del B
  del IT_companies


  #converting ages to a set and comparing the length of the list and the set
  age = set(ages)

  len_age = len(age)
  len_ages = len(ages)

  print(f"length of list of ages: {len_ages} Length of set of ages: {len_age}")

#main that will call all of the methods described in the ICP
#comment others out if you want to test them one at a time
def main():
  print(reversed())
  arithmetic()
  python_to_pythons()
  grading()
  type_of_elements()
  sets()

main()
```
